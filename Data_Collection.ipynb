```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
```

```python
num_users = 1000

usernames = [f"user_{i}" for i in range(num_users)]

base_df = pd.DataFrame({
    "username": usernames,
    "followers": np.random.randint(1000, 500000, num_users),
    "post_frequency": np.random.randint(1, 30, num_users)  # posts per month
})
```

```python
insta = base_df.copy()

insta["likes"] = insta["followers"] * np.random.uniform(0.05, 0.15, num_users)
insta["comments"] = insta["likes"] * np.random.uniform(0.02, 0.08, num_users)
insta["shares"] = insta["likes"] * np.random.uniform(0.01, 0.05, num_users)

insta["platform"] = "Instagram"
```

```python
twitter = base_df.copy()

twitter["likes"] = twitter["followers"] * np.random.uniform(0.01, 0.05, num_users)
twitter["comments"] = twitter["likes"] * np.random.uniform(0.01, 0.04, num_users)
twitter["shares"] = twitter["likes"] * np.random.uniform(0.05, 0.15, num_users)

twitter["platform"] = "Twitter"
```

```python
facebook = base_df.copy()

facebook["likes"] = facebook["followers"] * np.random.uniform(0.02, 0.08, num_users)
facebook["comments"] = facebook["likes"] * np.random.uniform(0.02, 0.06, num_users)
facebook["shares"] = facebook["likes"] * np.random.uniform(0.02, 0.10, num_users)

facebook["platform"] = "Facebook"
```

```python
df = pd.concat([insta, twitter, facebook], ignore_index=True)
```

```python
df["verified"] = np.random.choice([0, 1], size=len(df), p=[0.9, 0.1])

df["engagement_rate"] = (df["likes"] + df["comments"] + df["shares"]) / df["followers"]

df["is_fake"] = np.where(
    (df["engagement_rate"] < 0.01) & (df["followers"] > 10000),
    1,
    0
)
```

```python
# Select 3% rows
dirty_idx = df.sample(frac=0.03, random_state=42).index

# Introduce issues
df.loc[dirty_idx, "likes"] = None                 # missing values
df.loc[dirty_idx, "comments"] = -10              # invalid negative
df.loc[dirty_idx, "followers"] = 0               # invalid
df.loc[dirty_idx, "shares"] = df["shares"] * 50  # extreme outliers
```

```python
df["likes"] = df["likes"]
df["comments"] = df["comments"]
df["shares"] = df["shares"]
```
