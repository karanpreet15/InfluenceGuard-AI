# 📊 Exploratory Data Analysis (EDA)

This section analyzes influencer data after feature engineering to uncover:
- Engagement trends
- Platform performance
- Fake influencer patterns
- Feature relationships

##  Engagement Rate Distribution

This helps understand how engagement is distributed across influencers.

```python
sns.histplot(df["engagement_rate"], kde=True)
plt.title("Engagement Rate Distribution")
plt.show()
```

The engagement rate distribution is right-skewed, indicating that the majority of influencers have low to moderate engagement levels, while a small number of influencers achieve significantly higher engagement. This suggests that high engagement is rare and concentrated among a few top-performing influencers.

## Platform-wise Engagement Analysis

```python
df.groupby("platform")["engagement_rate"].mean().plot(kind="bar")
plt.title("Average Engagement Rate by Platform")
plt.show()
```

Instagram shows the highest engagement, followed by Facebook and Twitter. This suggests platform choice significantly impacts influencer performance.

##  Followers vs Engagement

```python
sns.scatterplot(x="followers", y="engagement_rate", data=df)
plt.title("Followers vs Engagement Rate")
plt.show()
```

The scatter plot shows a weak relationship between follower count and engagement rate, indicating that having more followers does not necessarily lead to higher engagement. Several accounts with high follower counts exhibit low engagement, suggesting potential fake or inactive followers.

## Correlation Analysis

```python
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()
```

## Top Influencers

```python
df.sort_values("engagement_rate", ascending=False)[
    ["username", "platform", "followers", "engagement_rate"]
].head(2000)
```

##  Fake Influencer Distribution

```python
sns.countplot(x="fake_flag", data=df)
plt.title("Fake vs Real Influencers")
plt.show()
```

A small percentage of influencers are flagged as fake, indicating anomalies in engagement patterns.

##  Viral Content Analysis

```python
sns.countplot(x="is_viral", data=df)
plt.title("Viral Content Distribution")
plt.show()
```

Only a small fraction of posts are viral, showing that high engagement is rare and significant.

##  Engagement Score Distribution

```python
sns.histplot(df["engagement_score"], kde=True)
plt.title("Engagement Score Distribution")
plt.show()
```

##  Key Insights Summary

- Most influencers have low engagement rates.
- Instagram performs better than other platforms.
- Weak correlation between followers and engagement suggests fake followers.
- Viral content is rare but impactful.
- Engagement score effectively captures influencer performance.
